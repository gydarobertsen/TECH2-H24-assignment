```python
#TECH2 mandatory assignment - Part B

#Importing packages I will use later in the program
import pandas as pd
from math import sqrt
import numpy as np

#Reading the data from the CSV_file using pandas
data = pd.read_csv('data.csv')

#Separating the data from the from the columns into lists, by using the iloc-function, and defining them as column 1,2 and 3
#Eliminating the NaN data using the dropna() function
#Adding the remaining data to lists separated in columns
column1 = data.iloc[:, 0].dropna().tolist()
column2 = data.iloc[:, 1].dropna().tolist()  
column3 = data.iloc[:, 2].dropna().tolist()
 
#Creating a function that computes the standard deviation using loops

def std_loops(column):
    sum1 = 0
    sum2 = 0
#Defining x as the length of the column
    x=len(column)
#Creating a for-loop to sum the values in the list together, and then summing the squared values of the list together
    for value in column:
        sum1 = sum1 + value
        sum2 = sum2 + value ** 2

#Computing the mean of the values, and of the squared values
    mean = sum1 / x
    S = sum2 / x
#Computing the variance
    var = S - mean ** 2
#Computing the standard deviation
    std = sqrt(var)

    return std

#Creating a function that computes the standard deviation using built-in functions
def std_builtin(column):
#Defining x to be the length of the column
    x=len(column)
#Converting the list into an array, so that I can perform mathematical procedures to each element
    column=np.array(column)
#Creating a list of the values from the array squared
    squared_column=column**2
#Calculating the mean
    mean2=sum(column)/x
#Calculating the mean of the squared values
    S2=sum(squared_column)/x
#Computing the variance
    var2=S2-mean2**2
#Computing the standard deviation 
    std2=sqrt(var2)
    
    return std2

#Creating a function that computes the standard deviation using python's proper function

def std(column):
    std3=np.std(column)
    return std3


#Timing all of the ouputs for column 1, using the %timeit function
time_loop_100= %timeit std_loops(column1)
time_builtin_100= %timeit std_builtin(column1)
time_std_100= %timeit std(column1)

#Timing all the outputs for column 2, using the %timeit function
time_loop_1000= %timeit std_loops(column2)
time_builtin_1000= %timeit std_builtin(column2)
time_std_1000= %timeit std(column2)

#Timing all the outputs for column 3, using the %timeit function
time_loop_10000= %timeit std_loops(column3)
time_builtin_10000= %timeit std_builtin(column3)
time_std_10000= %timeit std(column3)
      
```
